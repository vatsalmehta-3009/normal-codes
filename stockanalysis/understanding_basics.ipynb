```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# Resample, Shift, Rolling.
**resample:** used for aggregation through datetimeindex 
**shift   :** shifts whole data to the bottom n rows. 
**rolling :** for finding moving average, moving sum, etcetera.


```python
# loading tcs data
tcs_df = yf.download('TCS', 
                      start='2010-01-01', 
                      end='2021-06-12', 
                      progress=False)
tcs_df.head()
```

```python
tcs_df.resample(rule='A').mean().head()
```

```python
tcs_df.shift(-1)
```

```python
tcs_df['Open'].plot()
tcs_df.rolling(50).mean()['Open'].plot(figsize=(16,6)) 
# rolling: can calculate the moving averages.
```

```python
tcs_df['Open30MA'] = tcs_df.rolling(50).mean()['Open']
```

```python
tcs_df[['Open30MA','Open']].plot(figsize=(16,6))
```
