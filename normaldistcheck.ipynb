### Kolmogorov-Smirnov and Shapiro -Wilk tests(good reads)

```python
import scipy
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import statsmodels.api as sm
```

```python
# making data points for exponential function. y = e^x
shape, scale = 2., 10
popsize = 10000000
s = np.random.exponential(scale = scale,size=popsize)
# making samples to reach normal through CLT.
samplesize = 1000
samplecount = 1000
samples = np.array([np.mean(random.choices(s, k=samplesize)) for _ in range(samplecount)])
```

```python
sm.qqplot(samples, fit =True,line ='45')
plt.show()
#plt.show()
sns.distplot(samples)
plt.show()
```
