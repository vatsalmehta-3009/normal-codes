```python
# rcparams for decoration of graphs in matplot
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,20)
y1 = x ** 2
y2 = x ** 0.5
y3 = x
y4 = np.log10(x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
```

```python
# Making subplots
# 1. Count how mane are required
# 2. make number of rows and columns based on your requirement
```

```python
# (1,4),(4,1),(2,2)
plt.figure(figsize=(10,15))
plt.subplot(4,1,1)
plt.plot(x,y1,marker = 'd',c = 'k')
plt.subplot(4,1,2)
plt.plot(x,y2,marker = 'o',c = 'cyan')
plt.subplot(4,1,3)
plt.plot(x,y3,marker = 'o',c='yellow')
plt.subplot(4,1,4)
plt.plot(x,y4, marker = 'd', c = 'b')
```
