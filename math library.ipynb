```python
import math
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
x = float(input('Enter Angle in degrees: '))
print(x)
x = math.radians(x)
print(x)
print((math.sin(x))**2 + (math.cos(x))**2)
```

```python
plt.figure(figsize=(15,6))
interval = np.linspace(-90,90,10000)
rad_angles = list(map(math.radians,interval))
plt.plot(interval,list(map(math.sin,rad_angles)))
#plt.scatter(interval,list(map(math.sin,rad_angles)))
plt.plot(interval,list(map(math.cos,rad_angles)))
#plt.scatter(interval,list(map(math.cos,rad_angles)))
#plt.plot(interval,list(map(math.tan,rad_angles)))

plt.xticks(np.arange(min(interval), max(interval)+1, 15))
plt.plot(interval,[0 for i in interval])
plt.legend(['sine wave','sine dots','cosine wave','cosine dots','x-axis'])
plt.grid()
```
